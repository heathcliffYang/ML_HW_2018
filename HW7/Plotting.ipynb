```python
cl = []
for gamma in range(0, 10, 1):
    for data_num in range(2500, 5001, 2500):
        x = []
        y = []
        if (data_num == 2500):
            pca_file_name = "PCA_2d/PCA_test.txt"
        else:
            pca_file_name = "PCA_2d/PCA_train.txt"
        with open(pca_file_name, newline='') as csvfile:
            print("read "+pca_file_name)
            rows = csv.reader(csvfile)
            for row in rows:
                x.append(float(row[0]))
                y.append(float(row[1]))
        for kernel_choose in range(0,2,1):
            cl = []
            
            file_name = "ratio_cut_k_" + str(kernel_choose) + "_d_" + str(data_num) + "_g_" + str(gamma) + ".csv"
            print("read "+file_name)
            with open(file_name, newline='') as csvfile:
                rows = csv.reader(csvfile)
                for row in rows:
                    cl.append(int(row[0]))
                    
            axes = plt.gca()
            image_file_name = 'Ratio_cut_PCA_k_'+str(kernel_choose) + "_d_" + str(data_num)+ "_g_" +str(gamma)
            axes.set_title(image_file_name)
            for i in range(len(cl)):
                #print(x[i], y[i], cl[i])
                plt.scatter(x[i], y[i], c=color[cl[i]], marker='.')
            plt.savefig(image_file_name)
            #############################################
            file_name = "normalized_cut_k_" + str(kernel_choose) + "_d_" + str(data_num) + "_g_" + str(gamma) + ".csv"
            print("read "+file_name)
            with open(file_name, newline='') as csvfile:
                rows = csv.reader(csvfile)
                for row in rows:
                    cl.append(int(row[0]))
                    
            axes = plt.gca()
            image_file_name = 'Normalized_cut_PCA_k_'+str(kernel_choose) + "_d_" + str(data_num)+ "_g_" +str(gamma)
            axes.set_title(image_file_name)
            for i in range(len(cl)):
                #print(x[i], y[i], cl[i])
                plt.scatter(x[i], y[i], c=color[cl[i]], marker='.')
            plt.savefig(image_file_name)
```
