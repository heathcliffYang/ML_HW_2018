```python
import csv
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = []
y = []
with open("PCA_train.txt", newline='') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        x.append(float(row[0]))
        y.append(float(row[1]))
```

```python
cl = []
with open("T_train.csv", newline='') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        cl.append(int(row[0]))
```

```python
color = ['#da2614', '#b7e020', '#16e3c7', '#169fe3', '#2f5cf4', '#946cf1', '#f16ce5']
```

```python
## ground-truth
for i in range(len(cl)):
    plt.scatter(x[i], y[i], c=color[cl[i]-1], marker='.')
```
