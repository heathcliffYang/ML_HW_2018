```python
import csv
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = []
y = []
with open("PCA_test.txt", newline='') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        x.append(float(row[0]))
        y.append(float(row[1]))
```

```python
cl = []
with open("T_test.csv", newline='') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        cl.append(int(row[0]))
```

```python
color = ['#da2614', '#b7e020', '#f16ce5', '#169fe3', '#2f5cf4']
```

```python
## ground-truth
axes = plt.gca()
axes.set_title('PCA_test_gt')
for i in range(len(cl)):
    plt.scatter(x[i], y[i], c=color[cl[i]-1], marker='.')
plt.savefig('PCA_test_gt')
```
